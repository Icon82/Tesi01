```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carica i dati dal file CSV
df = pd.read_csv('data.csv')

# Seleziona i primi 4 valori
group1 = df[['Anthropogenic Forcing', 'Temperature Anomaly', 'CO2', 'CH4']]  # Modifica i nomi delle colonne in base al tuo file CSV

# Seleziona i secondi 5 valori
group2 = df[['Victims Without External Causes', 'Respiratory Diseases', 'Digestive Diseases', 'Mental and Behavioural Disorder', 'Nervous System Diseases']]  # Modifica i nomi delle colonne in base al tuo file CSV

# Calcola la matrice di correlazione per tutto il dataset
correlation_matrix = df.corr()

# Estrai le correlazioni specifiche tra i due gruppi
correlations = correlation_matrix.loc[group1.columns, group2.columns]

# Visualizza le correlazioni come una heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlazioni tra i primi 4 valori e gli altri 5")
plt.show()
```
